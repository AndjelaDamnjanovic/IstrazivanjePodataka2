```python
import pandas as pd
import csv
```

```python
data = pd.read_csv('datasetNew.csv')
data
```

```python
replaceOne=data.replace(1,'IND_1')
data=replaceOne.replace(0, 'IND_0')
data
```

```python
data_intersect=data[['repeat', 'intersect_ind']]
transactions=[]
items=[]
for index, row in data_intersect.iterrows():
    items = [str(x) for x in row.values]
    transactions.append(items)
```

```python
transactions
```

```python
#provera da li i dalje ima svih transakcija
print(len(transactions))
```

```python
file=open("transactions.csv", "w")
row_lists=[["repeat", "intersect_ind"]]
writer = csv.writer(file, delimiter=',')
for transaction in transactions:
    row_lists.append(transaction)
writer.writerows(row_lists)
file.close()
```
